```python

#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_profiling as pp


```

```python
#importing the datasets into python data frames

# The train and test datasets are loaded into Pandas dataframes and concatenated to one dataset

train_dataset = pd.read_csv('Train.txt')
test_dataset = pd.read_csv('Test.txt')

print(train_dataset.columns)
print('Train dataset : ' )
print(train_dataset.shape)

print(test_dataset.columns)
print('Test dataset : ' )
print(test_dataset.shape)

result = pd.concat([train_dataset,test_dataset])
print(result.columns)
print('Test dataset : ' )
print(result.shape)
```

```python
# Pandas Profile Report - Gives the details of each field mentioned in the data

pp.ProfileReport(result)
```

```python
# Making the output variable into Binomial Class : Attack or Normal
# Creating a new Field "Attack" (Note Capital 'A'ttack)

def attack_group(c):
  if c['attack'] == 'normal':
    return 'normal'
  else:
    return 'attack'

result['Attack'] = result.apply(attack_group, axis=1)
```

```python
# drop the 'attack' and last_flag fields

result.drop(['attack',' last_flag'],axis=1,inplace=True)

print(result.columns)
```

```python
# Label encoding for fields protocol_type , service & flag
# It is created as series
# Train & test data - Label encoded for the categorical field.

z_train = result.iloc[:, :-1].values

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_x_1 = LabelEncoder()
labelencoder_x_2 = LabelEncoder()
labelencoder_x_3 = LabelEncoder()
z_train[:, 1] = labelencoder_x_1.fit_transform(z_train[:, 1])
z_train[:, 2] = labelencoder_x_2.fit_transform(z_train[:, 2])
z_train[:, 3] = labelencoder_x_3.fit_transform(z_train[:, 3])
```

```python
# Assigning the series to the dataframe field
result['Protocol_type'] = z_train[:, 1]
result['Service'] = z_train[:, 2]
result['Flag']= z_train[:, 3]

for i in ['Protocol_type','Service','Flag']:
    
    result[i] = result[i].astype('int')
   
result.dtypes
```

```python
# After the fields are label encoded , drop the original fields 

result.drop(labels=['protocol_type','service','flag'],axis=1,inplace=True)
```

```python
print(result.columns)
```

```python
# defining the input variables X & output variable Y
# Choosing less fields as X below gives 99.37% accuracy in DT & 98.74% in Random Forest & Naive Beyes - 89%

#X=result[['duration','src_bytes','dst_bytes', 'logged_in', 'num_compromised','num_root','count','wrong_fragment','rerror_rate','dst_host_serror_rate', 'srv_rerror_rate','srv_diff_host_rate', 'dst_host_srv_diff_host_rate','dst_host_count','Service']]

# All the fields Except the eight fields that were "Rejected" in Pandas Profiling report are included in the input X to get more accuracy

X = result.drop(labels=['dst_host_rerror_rate','dst_host_serror_rate','dst_host_srv_rerror_rate','dst_host_srv_serror_rate','num_outbound_cmds','num_root','srv_rerror_rate','srv_serror_rate','Attack'],axis=1)
y = result['Attack']
```

```python
# comment this part when splitting is not required.

from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.grid_search import GridSearchCV

# Split the data to train & test - 70:30
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
# Use Decision Tree Alogorithm for classification problem

dt=DecisionTreeClassifier(criterion="entropy")

#dt=DecisionTreeClassifier(criterion="gini",max_depth=3,max_leaf_nodes=20)

dt_model=dt.fit(x_train,y_train)
```

```python
# Print the feature importance
dt_model.feature_importances_
list(zip(X,dt_model.feature_importances_))
```

```python
# print the accuracy score and confusion matrix - Decision Tree

dt_prd=dt_model.predict(x_test)
print(accuracy_score(y_test,dt_prd))
df_confusion = confusion_matrix(y_test,dt_prd)
df_confusion
```

```python
# Model with Random Forest

#x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)

rf_clf=RandomForestClassifier(max_depth = 10, min_samples_split=2, n_estimators = 10, random_state = 123)
rf_model=rf_clf.fit(x_train,y_train)
rf_prediction=rf_model.predict(x_test)

print(accuracy_score(y_test,rf_prediction))
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(x_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
print(accuracy_score(y_test,y_pred))
```
